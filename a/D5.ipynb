# Решение уравнения переноса
$u_{t}+\lambda u_{x}=0$


```python
from matplotlib import pyplot as plt
import numpy as np

lmb,L,T,dx,dt = 0.5,1,50,1e-3,1e-3 
sigma,N,M = lmb*dt/dx,int(T/dt),int(L/dx)
Grid = np.zeros((N, M))
Grid[0, int(M/10):int(M/10 * 2)] = Grid[0, int(M/10):int(M/10 * 2)] = 1

for i in range(2, N):
    for j in range(0, M):
        Grid[i, j] = 2*sigma*(2*sigma-1)/(sigma+1)*Grid[i-1, j-1]+2*sigma*Grid[i-2, j-1]+(2-4*sigma)*Grid[i-1, j]+(-2*sigma**2+3*sigma-1)/(sigma+1)*Grid[i-2, j]
        
plt.plot(Grid[0], label='t = 0')
plt.plot(Grid[200], label='t = 0.2')
plt.plot(Grid[400], label='t = 0.4')
plt.plot(Grid[600], label='t = 0.6')
plt.plot(Grid[4800], label='t = 48')
plt.plot(Grid[5000], label='t = 50')
plt.plot(Grid[5200], label='t = 52')
plt.legend()
plt.show()
```
